## 3. Baza Teorike
Formula për vëllimin e cilindrit:

$$ V = \pi r^2 h $$

Lidhja mes rrezes dhe diametrit:

$$ r = \frac{d}{2} $$

## 4. Aparatura dhe Procedura
### Aparatura
- Cilindër metalik  
- Pajisje matëse (kaliper)  
- Fletë për regjistrimin e të dhënave  

### Procedura
1. Vendoset cilindri në tavolinë.  
2. Matet diametri në tre pika.  
3. Regjistrohen rezultatet.

## 6. Analiza e të Dhënave
Mesatarja e diametrit:

$$ \bar{d} = \frac{10.1 + 10.2 + 9.9}{3} = 10.07 \ \text{mm} $$

Devijimi standard:

$$ \sigma = \sqrt{\frac{\sum (x_i - \bar{x})^2}{n-1}} $$

---

## 7. Analiza e Pasigurisë
- Pasiguria instrumentale: ±0.05 mm  
- Pasiguria statistikore: nga devijimi standard  

Propagimi i pasigurisë:

$$ \sigma_f^2 = \sum_i \left( \frac{\partial f}{\partial x_i} \sigma_{x_i} \right)^2 $$
