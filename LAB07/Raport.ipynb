## **Qëllimi i Laboratorit**

Ky laborator ka për qëllim të tregojë se dendësia nuk matet drejtpërdrejt, por llogaritet nga madhësi të tjera të matura:

-   Matja e **masës**
-   Matja e **dimensioneve gjeometrike**
-   Llogaritja e **vëllimit**
-   Përcaktimi i **dendësisë**
-   Analiza e **pasigurisë dhe përhapjes së saj**


## **Hyrje Teorike**

Dendësia përkufizohet si:

$$ \rho = \frac{m}{V} $$

Njësia SI: $$ \mathrm{kg/m^3} $$  
Njësia laboratorike: $$ \mathrm{g/cm^3} $$

Pasiguria relative për dendësinë:

$$ \frac{\Delta \rho}{\rho} \approx \frac{\Delta m}{m} + \frac{\Delta V}{V} $$


## **Pajisjet**

-   Peshore dixhitale (rezolucion 0.01 g)
-   Kalibër vernier (rezolucion 0.01 cm)
-   Objekt cilindrik (metal)

### **Pjesa A — Matja e Masës**

Matëm masën e cilindrit 5 herë:

| Matja | m (g) |
|-------|-------|
| 1     | 52.31 |
| 2     | 52.30 |
| 3     | 52.32 |
| 4     | 52.29 |
| 5     | 52.31 |

Mesatarja:  
$$ \bar{m} = 52.306 \, g $$  
Pasiguria: ±0.01 g


### **Pjesa B — Matja e Përmasave**

Matëm rrezen dhe lartësinë e cilindrit:

| Matja | r (cm) | h (cm) |
|-------|--------|--------|
| 1     | 1.50   | 5.00   |
| 2     | 1.51   | 5.01   |
| 3     | 1.50   | 5.00   |

Mesatarja:  
$$ r = 1.503 \, cm \quad h = 5.003 \, cm $$  
Pasiguria: ±0.01 cm

### **Pjesa C — Llogaritja e Vëllimit**

Formula për cilindrin:  
$$ V = \pi r^2 h $$

$$ V = \pi (1.503)^2 (5.003) \approx 35.5 \, cm^3 $$

Pasiguria relative:  
$$ \frac{\Delta V}{V} \approx 2\frac{\Delta r}{r} + \frac{\Delta h}{h} $$

$$ \frac{\Delta V}{V} \approx 2\cdot\frac{0.01}{1.503} + \frac{0.01}{5.003} \approx 0.014 $$

Pra:  
$$ \Delta V \approx 0.014 \cdot 35.5 \approx 0.50 \, cm^3 $$

### **Pjesa D — Llogaritja e Dendësisë**

$$ \rho = \frac{m}{V} = \frac{52.306}{35.5} \approx 1.47 \, g/cm^3 $$

Pasiguria relative:  
$$ \frac{\Delta \rho}{\rho} \approx \frac{\Delta m}{m} + \frac{\Delta V}{V} $$

$$ \frac{\Delta \rho}{\rho} \approx \frac{0.01}{52.306} + 0.014 \approx 0.0142 $$

Pasiguria absolute:  
$$ \Delta \rho = 1.47 \cdot 0.0142 \approx 0.02 \, g/cm^3 $$

## **Rezultati Final**

$$ \rho = (1.47 \pm 0.02) \, g/cm^3 $$

## **Diskutim**

-   Pasiguria dominohet nga matja e dimensioneve (rrezja dhe lartësia).
-   Masa është shumë më e saktë krahasuar me përmasat.
-   Rezultati është afër dendësisë së aluminit (2.7 g/cm^3), por më i ulët → objekti mund të jetë aliazh ose material tjetër.

## **Mesazhi Kryesor**

Ky laborator tregon se:

-   Dendësia është një madhësi **indirekte**
-   Pasiguria e saj varet nga pasiguritë e matjeve bazë
-   Analiza e pasigurisë është pjesë e domosdoshme e interpretimit fizik
