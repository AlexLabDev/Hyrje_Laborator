## 1. Pse përdorim Python në laborator?

Në laborator shpesh duhet të:

-   analizojmë **rezultate nga eksperimente fizike**
-   ndërtojmë **grafikë krahasuese**
-   përshtatim modele matematikore për të dhëna reale
-   nxjerrim **përfundime numerike**

Python është i dobishëm sepse:

-   është **i thjeshtë për t’u përdorur**

-   përdoret gjerësisht në **fizikë, inxhinieri dhe biologji**

Bibliotekat kryesore që do të përdorim:

| Biblioteka     | Qëllimi                           |

| **NumPy**      | llogaritje numerike dhe vektorë   |
