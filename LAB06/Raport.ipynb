## **Qëllimi i Laboratorit**

Ky laborator ka për qëllim të ndërtojë bazën e matjes eksperimentale përmes:

-   Matjes së **gjatësisë, masës dhe kohës**
-   Kuptimit të **pasigurisë dhe precizionit**
-   Identifikimit të **gabimeve sistematike dhe rastësore**
-   Regjistrimit dhe analizës së të dhënave në mënyrë shkencore

## **Konceptet Kryesore**

### 1. Matja si proces fizik

Çdo matje është një krahasim me një njësi standarde dhe **gjithmonë përmban pasiguri**.

$$ x = x_{\text{mat}} \pm \Delta x $$

### 2. Precizioni dhe rezolucioni

-   **Rezolucioni**: ndarja më e vogël e instrumentit
-   **Precizioni**: sa afër janë matjet mes tyre

Rregull praktik:
Pasiguria instrumentale ≈ **gjysma e ndarjes më të vogël**

### 3. Llojet e gabimeve

-   **Sistematik** → zhvendosje konstante (kalibrim i gabuar)
-   **Rastësor** → shpërndarje statistikore
-   **Trashanik** → gabime njerëzore (duhen eliminuar)

## **Pjesa A — Matja e Gjatësisë**

### **Procedura**

Matëm gjatësinë e një kubi druri me vizore, kalibër dhe mikrometër. Për secilin instrument kryem 5 matje.

### **Të dhënat**

| Instrument | Matja 1 | Matja 2 | Matja 3 | Matja 4 | Matja 5 | Mesatarja | Pasiguria |
|------------|---------|---------|---------|---------|---------|-----------|-----------|
| Vizore     | 20.1 mm | 20.0 mm | 20.2 mm | 20.1 mm | 20.0 mm | 20.08 mm  | ±0.5 mm   |
| Kalibër    | 20.05 mm| 20.04 mm| 20.06 mm| 20.05 mm| 20.04 mm| 20.048 mm | ±0.02 mm  |
| Mikrometër | 20.052 mm|20.051 mm|20.053 mm|20.052 mm|20.051 mm| 20.052 mm | ±0.005 mm |

### **Analiza**

-   Mesatarja me vizore: $$ \bar{x} = 20.08 \, \text{mm} $$
-   Mesatarja me kalibër: $$ \bar{x} = 20.048 \, \text{mm} $$
-   Mesatarja me mikrometër: $$ \bar{x} = 20.052 \, \text{mm} $$

**Përfundim**: Mikrometri është më preciz, vizorja ka pasiguri më të madhe.



## **Pjesa B — Matja e Masës**

### **Procedura**

Matëm masën e kubit me peshore dixhitale, 5 herë.

### **Të dhënat**

| Matja 1 | Matja 2 | Matja 3 | Matja 4 | Matja 5 | Mesatarja | Pasiguria |
|---------|---------|---------|---------|---------|-----------|-----------|
| 15.02 g | 15.01 g | 15.03 g | 15.02 g | 15.01 g | 15.018 g  | ±0.01 g   |

### **Analiza**

-   Mesatarja: $$ \bar{m} = 15.018 \, \text{g} $$
-   Pasiguria: ±0.01 g
-   Nuk u vërejt drift i dukshëm, rezultatet janë konsistente.

## **Pjesa C — Matja e Kohës**

### **Procedura**

Matëm kohën e rënies së një topi ping-pong nga lartësia 1.0 m. Kryem 10 matje me kronometër manual.

### **Të dhënat**

| Matja 1 | Matja 2 | Matja 3 | Matja 4 | Matja 5 | Matja 6 | Matja 7 | Matja 8 | Matja 9 | Matja 10 |
|---------|---------|---------|---------|---------|---------|---------|---------|---------|----------|
| 0.45 s  | 0.47 s  | 0.46 s  | 0.44 s  | 0.46 s  | 0.45 s  | 0.47 s  | 0.46 s  | 0.45 s  | 0.46 s   |

### **Analiza**

-   Mesatarja: $$ \bar{t} = 0.456 \, \text{s} $$
-   Devijimi standard ≈ 0.01 s
-   Pasiguria kryesore vjen nga **reagimi njerëzor** (start/stop manual).

**Krahasim teorik**:  
Nga formula $$ t = \sqrt{\frac{2h}{g}} $$  
me $$ h = 1.0 \, \text{m}, g = 9.81 \, \text{m/s}^2 $$  
del $$ t_{\text{teorik}} = 0.45 \, \text{s} $$  

→ Rezultati eksperimental përputhet mirë me teorinë.
