## 1. Përshkrimi i metodës
Të dhënat u gjeneruan përmes skriptit `generate_data.py` duke përdorur **GitHub ID** unik.  
Ky identifikues siguron që dataset-i të jetë **unik** dhe **i riprodhueshëm**.  
Parametrat e modelit linear (pjerresia *a* dhe prerja *b*) si dhe zhurma e shtuar janë të lidhura me këtë ID, duke garantuar integritetin akademik.  

Pas gjenerimit, file-i `model.dat` u vendos në strukturën e kërkuar (`LAB05/data/`) dhe u ngarkua përmes `numpy`.

---

## 2. Ngarkimi i të dhënave
Në këtë hap, të dhënat e gjeneruara ruhen në file-in `model.dat` dhe ngarkohen në program përmes bibliotekës `numpy`.  
Ky proces siguron që të kemi vektorët **x_Data** dhe **y_Data** të cilët përfaqësojnë variablat eksperimentale.

```python
import numpy as np
import matplotlib.pyplot as plt

# Lexo direkt nga i njëjti folder ku ndodhet notebook-u
data = np.loadtxt("model.dat", delimiter=",", skiprows=1)

x_Data = data[:, 0]
y_Data = data[:, 1]
```

## 3. Përshtatja lineare
Në këtë hap kryhet përshtatja lineare e të dhënave duke përdorur funksionin `polyfit`.  
Rezultati është një vijë e drejtë që përshkruan lidhjen mes variablave **x** dhe **y**.  
Parametrat e fit-it janë pjerresia (*a*) dhe prerja (*b*).

Shembull:

Pjerresia e drejtezes a =  1.85
Prerja me boshtin b =  0.42

```python
a, b = np.polyfit(x_Data, y_Data, 1)

print(f"Pjerresia e drejtezes a = {a : .2f}")
print(f"Prerja me boshtin b = {b : .2f}")
```

## 4. Grafiku i të dhënave dhe fit-it
Në këtë hap vizualizohen të dhënat eksperimentale së bashku me vijën e regresionit linear.  
Grafiku paraqet pikat e dataset-it dhe vijën e kuqe të modelit linear, duke treguar përshtatjen e bërë me funksionin `polyfit`.

```python
xfit = np.linspace(1,10,100)
yfit = a*xfit + b

plt.figure(1,(8,6))
plt.plot(x_Data, y_Data, 'ok', label='Te Dhenat')
plt.plot(xfit, yfit, '-r', lw=2, label=f'y={a: .2f}*x + {b: .2f}')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## 5. Grafiku i mbetjeve (residuals)
Mbetjet tregojnë diferencën mes vlerave të matura dhe modelit linear.  
Nëse shpërndahen rreth zeros pa një trend të dukshëm, modeli linear konsiderohet i përshtatshëm.  
Ky hap ndihmon në vlerësimin e cilësisë së përshtatjes.

```python
# 5. Grafiku i mbetjeve (residuals)
plt.figure(2,(8,6))
plt.stem(x_Data, y_Data - (a*x_Data + b), '-r')
plt.hlines(0,0,11,ls='--',color='k',lw=1)
plt.xlabel("x")
plt.ylabel("mbetjet")
plt.xlim(0.8,10.2)
plt.show()
```

## 6. Interpretimi fizik/statistik i rezultateve
Në këtë hap bëhet analiza e rezultateve të marra nga përshtatja lineare.  
Interpretimi ndihmon të kuptojmë kuptimin fizik dhe statistikor të parametrave të modelit.

- **Modeli linear** përshkruan lidhjen mes variablave **x** dhe **y**.  
- **Parametri a (pjerresia)** → tregon intensitetin e lidhjes: sa rritet y për çdo njësi të x.  
- **Parametri b (prerja)** → jep vlerën fillestare të y kur x = 0.  
- **Mbetjet** → tregojnë cilësinë e përshtatjes.  
  - Nëse shpërndahen rreth zeros pa një trend të dukshëm, modeli linear është i përshtatshëm.  
  - Nëse shfaqin strukturë sistematike, kjo sugjeron se një model më kompleks mund të nevojitet.

Ky interpretim siguron që rezultatet të mos mbeten vetëm matematikore, por të kenë kuptim praktik dhe statistikor në analizën e të dhënave.
